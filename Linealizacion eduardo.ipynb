```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def sys1(X, t):
    x1, x2 = X
    m = 1
    b = 1
    R = 1
    g = 9.8
    dx1 = x2
    dx2 = -(1/(m * R ** 2)) * x2 - (g / R) * np.sin(x1)
    return [dx1, dx2]

x0 = [3, 0]
t = np.linspace(0, 20, 400)

sol = odeint(sys1, x0, t)

plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

$$ \large  \dot V(x) = x (x^2 + u)  $$

$$ \large  \dot V(x) = x^3 + xu  $$

$$ \large  \dot V(x) = x^3 + x(-2x^2)  $$

```python
def f2(x,t):
    u = -2*x**2
    dx1 = x**2 + u 
    return dx1


t=np.arange(0,3,0.1)
x0= [2,1]

ys = odeint(f2,x0,t)


plt.plot(t,ys[:,0], 'b', label='x1(t)')
plt.plot(t,ys[:,1], 'g', label='x2(t)')
plt.legend(loc='best')
plt.show()
```
